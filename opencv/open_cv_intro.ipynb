```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = "people.jpg"
img = cv2.imread(img_path)
print(img.shape)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
face_detections = face_detector.detectMultiScale(img_gray, scaleFactor=1.01, minSize=(180,180), minNeighbors=8, maxSize=(800,800))
```

```python
face_detections
```

```python
eye_detector = cv2.CascadeClassifier('haarcascade_eye.xml')
eye_detections = eye_detector.detectMultiScale(img_gray, scaleFactor=1.08, minNeighbors=10, minSize=(80,80))
```

```python
for (x,y,w,h) in face_detections:
    cv2.rectangle(img, (x,y), (x + w, y + h), (0,255,0), 2)

for (x,y,w,h) in eye_detections:
    cv2.rectangle(img, (x,y), (x + w, y + h), (0,0,255), 2)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
